```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# Raw data file
file_to_load = "Resources/purchase_data.csv"
```

```python
# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

```python
# Player Count
# Display the total number of players
unique_players = purchase_data["SN"].unique()
unique_players 
```

```python
total_players = pd.DataFrame(unique_players).count()
totalplayers = pd.DataFrame(total_players, columns = ['Total Players'])
totalplayers
```

```python
# Purchasing Analysis (Total)
# Run basic calculations to obtain number of unique items, average price, etc.
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame
average_price = purchase_data["Price"].mean()
average_price
```

```python
number_of_purchase = purchase_data["Purchase ID"].value_counts()
sum_num_purchase = number_of_purchase.sum()
sum_num_purchase
```

```python
unique_items = purchase_data["Item Name"].unique()
count_of_unique_items = pd.DataFrame(unique_items).count()
count_of_unique_items
```

```python
total_revenue = purchase_data["Price"].sum()
total_revenue
```

```python
purchase_analysis = {'Average Price': average_price,
                    'Number of Purchases': sum_num_purchase,
                    'Number of Unique Items': count_of_unique_items,
                    'Total Revenue': total_revenue}
purchase_analysis_df = pd.DataFrame(purchase_analysis)
purchase_analysis_df
```

```python
# Gender Demographics
# Percentage and Count of Male Players
# Percentage and Count of Female Players
# Percentage and Count of Other / Non-Disclosed
total_players = purchase_data['Gender'].count()
total_players
```

```python
gender_demo = {'Gender': ['Male', 'Female', 'Other'], 'Total Count': [male_players, female_players, other_players], 'Percentage of Players': [male_percentage,female_percentage, other_percentage]}
gender_demo
```

```python
gender_demo_df = pd.DataFrame(gender_demo,columns=['Gender', 'Total Count', 'value'])
gender_index = gender_demo_df.set_index('Gender')
gender_index
```

```python
# Purchasing Analysis (Gender)
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame
genders = purchase_data["Gender"].unique()
genders
```

```python
purchase_count = purchase_data["Purchase ID"].sum()
purchase_count
```

```python
gender_purchases = purchase_data.groupby('Gender').agg({'Purchase ID':'count'})
gender_purchases
```

```python
average_purchases = purchase_data.groupby('Gender').agg({'Price':'mean'})
average_purchases_format = average_purchases.applymap("${0:.2f}".format)
average_purchases_format
```

```python
total_purchases = purchase_data.groupby('Gender').agg({'Price':'sum'})
total_purchases_format = total_purchases.applymap("${0:.2f}".format)
total_purchases_format
```

```python
sn = purchase_data.groupby(['SN','Gender']).agg({'Price':'sum'})
sn_gender = sn.groupby('Gender').agg({'Price':'mean'})
sn_gender_format = sn_gender.applymap("${0:.2f}".format)
sn_gender_format
```

```python
purchasing_gender_analysis_2 = pd.concat([gender_purchases, average_purchases_format,total_purchases_format, sn_gender_format], axis=1)
purchasing_gender_analysis_2
```

```python
purchasing_gender_analysis_2.columns = ['Purchase Count', 'Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']
purchasing_gender_analysis_2
```

```python
# Age Demographics
# Establish bins for ages
# Categorize the existing players using the age bins. Hint: use pd.cut()
# Calculate the numbers and percentages by age group
# Create a summary data frame to hold the results
# Optional: round the percentage column to two decimal points
# Display Age Demographics Table
# Establish bins for ages
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_bin = pd.cut(purchase_data["Age"], bins, labels=group_names)
age_bin
```

```python
age_bins_df = pd.DataFrame(age_bin)
age_bins_df
```

```python
age_group = age_bins_df.groupby("Age")
age_group_count = age_group["Age"].count()
age_group_count
```

```python
age_group_count_df = pd.DataFrame(age_group_count)
age_group_count_df
```

```python
age_group_count_df.columns = ['Total Count']
age_group_count_df
```

```python
age_group_count_df['Total Count']
```

```python
age_group_count_df["Percentage of Players"] = age_group_count_df["Total Count"]/age_group_count_df["Total Count"].sum()*100
age_group_count_df
```

```python
# Purchasing Analysis (Age)
# Bin the purchase_data data frame by age
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame
purchase_data.head()
```

```python
purchase_data["Ranges"] = pd.cut(purchase_data["Age"], bins, labels = group_names)
purchase_data
```

```python
age_purchase_data = purchase_data.groupby(['Ranges'])

purchase_count = age_purchase_data['Purchase ID'].count()
purchase_count
```

```python
total_purchase = age_purchase_data['Price'].sum()
total_purchase
```

```python
players = purchase_data.drop_duplicates(subset ='SN').groupby(['Ranges'])
players = players['SN'].count()
players
```

```python
purchase_age = pd.DataFrame(age_purchase_data['Purchase ID'].count())
purchase_age
```

```python
purchase_age = purchase_age.rename(columns = {'Purchase ID':'Purchase Count'})
purchase_age
```

```python
purchase_age['Average Purchase Price'] = round((total_purchase/purchase_count),2)
purchase_age
```

```python
purchase_age['Total Purchase Value'] = total_purchase
purchase_age
```

```python
purchase_age['Avg Total Purchase Per Person'] = round((total_purchase/players),2)
purchase_age
```

```python
# Top Spenders
# Run basic calculations to obtain the results in the table below
# Create a summary data frame to hold the results
# Sort the total purchase value column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame
purchase_data
```

```python
grouped_by_sn_final.columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']
grouped_by_sn_final
```
